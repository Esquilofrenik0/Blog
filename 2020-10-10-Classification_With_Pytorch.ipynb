### Classification 

In machine learning, classification refers to a predictive modeling problem where a class label is predicted for a given example of input data. 

A classification model tries to draw some conclusion from the input values given for training. It will predict the class labels/categories for the new data. 

In 2 dimension and with only 2 labels, we can define the solution to a classification problem as the line that separates the points of the 1st class from the points of the 2nd class.

Here we can visualize a simple classification problem where we classify if the x1 value of a given point is positive or negative:

```python
sepx = [0,0]
sepy = [-1,1]
X = torch.rand(20,2).type(torch.FloatTensor)*2-1
Y = torch.empty(20)

for i in range(len(X)):
    if(X[i][0]>=0):
        Y[i] = 1
    else:
        Y[i] = 0
        
plt.xlabel('x1 value')
plt.ylabel('x2 value')
plt.plot(sepx,sepy)
plt.scatter(X[:, 0], X[:, 1],c=Y)
```

We can see here that, for any given point, if the 1st value is less than 0, it is classified as negative and vice-versa.

In this case it is easy enough for a human to create a function which draws the desired line to solve our classification problem.

Now imagine that we wanted to separate the points into 3 classes, class 0 if both the x1 value and x2 value are positive, class 1 if both the x1 value and the x2 value are negative and class 2 for the other properties.

```python
for i in range(len(X)):
    if X[i][0]>=0 and X[i][1]>=0:
        Y[i] = 0
    elif X[i][0]<0 and X[i][1]<0:
        Y[i] = 1
    else:
        Y[i] = 2
        
plt.xlabel('x1 value')
plt.ylabel('x2 value')
plt.scatter(X[:, 0], X[:, 1],c=Y)
```

Despite this being a simple algorithm to program using if statements, it is now much harder to describe a function that draws a line that separates these 3 classes.

By using Neural Networks we can draw this line without having to create a new function.
